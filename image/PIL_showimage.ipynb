```python
#显示灰度图像，
import os
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(os.path.join('/home/ginchoeng/samples/all_data/train/320.jpg'))
img = img.convert('L')  #多添加了这一行,因为所有图片读取了之后都是RGB格式，但是灰度图像的格式是'L'，所以就要经过这一行的转换

plt.figure('Image')
plt.imshow(img,cmap='gray')  #在读取图片时，多添加后面这个cmap参数,这是一定要添加的 cmap: 颜色图谱（colormap), 默认绘制为RGB(A)颜色空间。
plt.axis('on')
plt.title('bin')
plt.show()
```

```python
#一个窗口显示多个图片
%matplotlib inline
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


img = Image.open(os.path.join('/home/ginchoeng/samples/all_data/train/320.jpg'))
gray = img.convert('L')
r,g,b = img.split() #这一行的意思是，分离图片当中的 3个通道
img_merged = Image.merge('RGB',(r,g,b)) #合并这三个通道


plt.figure(figsize=(10,5))#设置窗口的大小
plt.suptitle('mul_image')
plt.subplot(2,3,1),plt.title('image')
plt.imshow(img),plt.axis('on')

plt.subplot(2,3,2),plt.title('gray')
plt.imshow(gray,cmap='gray'),plt.axis('on')

plt.subplot(2,3,3),plt.title('merged')
plt.imshow(img_merged),plt.axis('on')

plt.subplot(2,3,4),plt.title('r')
plt.imshow(r,cmap='gray'),plt.axis('on')

plt.subplot(2,3,5),plt.title('g')
plt.imshow(g,cmap='gray'),plt.axis('on')

plt.subplot(2,3,6),plt.title('b')
plt.imshow(b,cmap='gray'),plt.axis('on')

plt.show()
```
