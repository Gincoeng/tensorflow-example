```python
#####使用opencv读取图片,得到的就是数组
import cv2

path = '/home/ginchoeng/samples/1.jpg'
img = cv2.imread(path)
print(type(img))     #这里得到的是这张图片的数组



####使用opencv把数组保存为图片
cv2.imwrite('0000.jpg',img)
```

```python
####使用matplotlib打开图片，跟CV2是一样的，都是数组
import matplotlib.image as mi

path = '/home/ginchoeng/samples/1.jpg'
img = mi.imread(path)
print(type(img))


###使用matplotlib把数组保存为图片
mi.imsave('0000.jpg',img)
```

```python
#使用pillow读取图片，得到的不是数组
from PIL import Image

path = '/home/ginchoeng/samples/1.jpg'
img = Image.open(path)
print(type(img))
print(img)

#使用pillow保存图片
img.save('0000.jpg')
```

```python
#使用matplotlib保存图片的另一种方法
import matplotlib.pyplot as plt

data = [1,5,2,9,4,9,1,7,3,8]
plt.plot(data,color='g')
plt.scatter(list(range(10)),data,color='r')
plt.savefig('0.jpg')
```
