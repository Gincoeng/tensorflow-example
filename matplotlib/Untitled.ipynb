```python
#生成填充图形
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.axes([0.025,0.025,0.95,0.95])   #[xmin,ymin,xmax,ymax] 设定范围

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.fill_between(X, 1, Y+1, color='blue', alpha=.25)

plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.fill_between(X, -1, Y-1, (Y-1) > -1, color='blue', alpha=.25)
plt.fill_between(X, -1, Y-1, (Y-1) < -1, color='red',  alpha=.25)

plt.plot(X,Y,'r')   #这是想显示出原来的曲线y

plt.xlim(-np.pi,np.pi), plt.xticks([])  #设置X轴的上下限，  以及X轴的标记    tickes中不填写就是默认为 none
plt.ylim(-2.5,2.5), plt.yticks([])      #设置y轴的上下限，  以及y轴的标记    tickes中不填写就是默认为 none
# savefig('../figures/plot_ex.png',dpi=48)
plt.show()
```

```python
#生成散点图
import numpy as np
import matplotlib.pyplot as plt

n = 1024
X = np.random.normal(0,1,n) #均值为0，标准差为1，维度是n的正态分布矩阵
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X) #按照元素(x2,x1)所在象限返回x1/x2的反正切值。    在这里x对应x1   y对应x2

plt.axes([0.025,0.025,0.95,0.95])  #[xmin,ymin,xmax,ymax]设置是x，y轴的上下限
plt.scatter(X,Y, s=75, c=T, alpha=.5)    #画出散点图  x为横轴，y为纵轴，s代表图中点的大小。c代表颜色，alpha代表透明度

plt.xlim(-1.5,1.5), plt.xticks([])
plt.ylim(-1.5,1.5), plt.yticks([])
# savefig('../figures/scatter_ex.png',dpi=48)
plt.show()
```

```python
#生成条形图
import numpy as np
import matplotlib.pyplot as plt

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)   #float就是浮点数，后边的np.random.uniform  是在0.5到1之间随机输出n个数
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.axes([0.025,0.025,0.95,0.95])   #[xmin,ymin,xmax,ymax]设置是x，y轴的上下限
plt.bar(X, +Y1, facecolor='g', edgecolor='r')         #画出条形图，前面两个参数是x，y。第三个是 条形 的颜色  第四个是 条形边缘 的颜色
plt.bar(X, -Y2, facecolor='m', edgecolor='y')         #画出条形图

# x以及y后面跟着的是数字，是代表着得出的数字应该标在什么位置，然后的参数是 2位小数，ha是横向居中对齐，va是底部对齐

for x,y in zip(X,Y1):
    plt.text(x+0.2, y+0.05, '%.2f' % y, ha='center', va= 'bottom')  #plt。text是在相应位置增加文本

for x,y in zip(X,Y2):
    plt.text(x+0.2, -y-0.05, '%.2f' % y, ha='center', va= 'top')

    
    
plt.xlim(-.5,n), plt.xticks([])
plt.ylim(-1.25,+1.25), plt.yticks([])

# savefig('../figures/bar_ex.png', dpi=48)
plt.show()
```

```python
#生成等高线图
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):                                 #这是用来生成高度的函数
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)   #在二维平面中将每一个x和每一个y分别对应起来，编织成栅格

plt.axes([0.025,0.025,0.95,0.95])   #[xmin,ymin,xmax,ymax]设置是x，y轴的上下限

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap=plt.cm.hot)  #这是进行颜色填充。
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5) #画等高线   这里的8 是代表把等高线分成几部分
plt.clabel(C, inline=1, fontsize=10)  #绘制等高线数据

plt.xticks([]), plt.yticks([])
# savefig('../figures/contour_ex.png',dpi=48)
plt.show()
```

```python
#Imshow
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,3,3.5*n)
y = np.linspace(-3,3,3.0*n)
X,Y = np.meshgrid(x,y)    #在二维平面中将每一个x和每一个y分别对应起来，编织成栅格
Z = f(X,Y)

plt.axes([0.025,0.025,0.95,0.95])  #[xmin,ymin,xmax,ymax]设置是x，y轴的上下限
plt.imshow(Z,interpolation='nearest', cmap='bone', origin='lower')  #绘制Z的图像， 第二个参数是 间隔明确 ，cmap就是颜色 ，orign 就是绘制方向
plt.colorbar(shrink=.92)   #shrink就是所占图片的颜色比例

plt.xticks([]), plt.yticks([])
# savefig('../figures/imshow_ex.png', dpi=48)
plt.show()
```

```python
#饼图
import numpy as np
import matplotlib.pyplot as plt

n = 20
Z = np.ones(n)   #生成全一矩阵
Z[-1] *= 2       #最后一列设为2

plt.axes([0.025,0.025,0.95,0.95])   #[xmin,ymin,xmax,ymax]设置是x，y轴的上下限

plt.pie(Z, explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])
# #画图,参数：传进的数组；explode:每个扇形离中心的距离;colors:设置颜色(每一个都有不同的颜色)这里用的灰度颜色，也可以使用rgb;

plt.gca().set_aspect('equal')   #生成一个正圆不是一个椭圆
plt.xticks([]), plt.yticks([])

# savefig('../figures/pie_ex.png',dpi=48)
plt.show()
```

```python
#矢量图
import numpy as np
import matplotlib.pyplot as plt

n = 8
X,Y = np.mgrid[0:n,0:n]
T = np.arctan2(Y-n/2.0, X-n/2.0)
R = 10+np.sqrt((Y-n/2.0)**2+(X-n/2.0)**2)
U,V = R*np.cos(T), R*np.sin(T)

plt.axes([0.025,0.025,0.95,0.95])
plt.quiver(X,Y,U,V,R, alpha=.5)
plt.quiver(X,Y,U,V, edgecolor='k', facecolor='None', linewidth=.5)

plt.xlim(-1,n), plt.xticks([])
plt.ylim(-1,n), plt.yticks([])

# savefig('../figures/quiver_ex.png',dpi=48)
plt.show()
```

```python
#极轴图
import numpy as np
import matplotlib.pyplot as plt

ax = plt.axes([0.025,0.025,0.95,0.95], polar=True)  #polar就是开启极轴模式

N = 20
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
radii = 10*np.random.rand(N)   #创建一个随机数组。扩大10倍。因为生成的数都是0-1之间
width = np.pi/4*np.random.rand(N)   #创建一个随机数组
bars = plt.bar(theta, radii, width=width, bottom=0.0)   #画的条形图的形状设定

for r,bar in zip(radii, bars):
    bar.set_facecolor( plt.cm.jet(r/10.))  #设定颜色
    bar.set_alpha(0.5)                     #设定透明度

ax.set_xticklabels([])
ax.set_yticklabels([])
# savefig('../figures/polar_ex.png',dpi=48)
plt.show()
```

```python
#3D
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()   #定义一个窗口，
ax = Axes3D(fig)     #在这个窗口添加上一个3D坐标
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)   ##在二维平面中将每一个x和每一个y分别对应起来，编织成栅格

R = np.sqrt(X**2 + Y**2)  #本行以及一下这行，计算（x，y）点的高度值
Z = np.sin(R)             #

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)  #第4，5个参数代表row跟column的跨度，最后一个参数是颜色
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)  #添加xy平面的等高线， 如果zdir选择了x 效果就相当于对xz平面投影。  画在z=-2的平面上
ax.set_zlim(-2,2)  #设置z轴的上下限

# savefig('../figures/plot3d_ex.png',dpi=48)
plt.show()
```
